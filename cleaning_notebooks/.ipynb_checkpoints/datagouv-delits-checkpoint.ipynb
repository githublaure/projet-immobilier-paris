```python
def read_files(path, sheet):
    df = pd.read_excel(path, sheet_name = sheet, header=2)
    columns_to_keep = [x for x in df.columns.values if any(w in x for w in ['ARRONDISSEMENT', 'PARIS CENTRE', 'DTSP92', 'DTSP93', 'DTSP94'])]
    
    nm_columns = list(df.columns)
    nm_columns.pop(1)
    nm_columns.insert(1, 'c1')
    
    df.columns = nm_columns
    
    df = df.query("c1 not in ['Autres vols simples contre des particuliers dans des locaux ou lieux publics', 'Vols à la tire']")
    
    df = df.loc[:, columns_to_keep].sum(axis=0).to_frame()
    
    if "DTSP75 01ER ARRONDISSEMENT" in df.index.values:
        df.rename(index={'DTSP75 01ER ARRONDISSEMENT':'DTSP75 PARIS CENTRE'})
    return df.to_dict()
```

```python
path = '../data/crimes-et-delits-enregistres-par-les-services-de-gendarmerie-et-de-police-depuis-2012(1).xlsx'

df_dict = {}

for i in range(2016, 2021):
    sheet = "Services PN " + str(i)
    df_dict[i] = read_files(path, sheet)
```

```python
df = None
for i, v in df_dict.items():
    if df is None:
        df = pd.DataFrame(v).reset_index()
        df.columns = ['ids', str(i)]
    else:
        ndf = pd.DataFrame(v).reset_index()
        ndf.columns = ['ids', str(i)]
        df = pd.merge(df, ndf, left_on = 'ids', right_on='ids', how='outer')
        
```

```python
df.columns = ['ids', 'y16', 'y17', 'y18', 'y19', 'y20']
```

```python
df1 = (df
 .fillna({'ids':''})
 .assign(ids=lambda x: np.where(x.ids=='DTSP75 PARIS CENTRE', 'DTSP75 01ER ARRONDISSEMENT', x.ids))
 .groupby('ids', as_index=False)
 .agg(y16=('y16','sum'), y17=('y17', 'sum'), y18=('y18', 'sum'), y19=('y19', 'sum'), y20=('y20', 'sum'))
 .assign(ids=lambda x: x.ids.astype('str'))
 .assign(ids=lambda x: x.ids.apply(lambda y: re.sub(r'DTSP75 (\d+)', r'750\1', y)))
 .assign(ids=lambda x: x.ids.str.replace('DTS', ''))
 .assign(ids = lambda x: np.where(x.ids=='75006EME ARRONDISSEMENT', '75005EME ET 06EME ARRONDISSEMENTS', x.ids))
 .groupby('ids', as_index=False)
 .agg(y16=('y16','sum'), y17=('y17', 'sum'), y18=('y18', 'sum'), y19=('y19', 'sum'), y20=('y20', 'sum'))
 .assign(departement=lambda x: x.ids.apply(lambda y: re.search('75|92|93|94', y).group(0)))
 .assign(code_postal=lambda x: x.ids.apply(lambda y: re.search('\d{5}|\d+', y).group(0)))
 .assign(ville=lambda x: x.ids.apply(lambda y: re.sub('P\d+|\d+ER|\d+EME', '', y)))
 .assign(ville=lambda x: x.ville.apply(lambda y: re.sub('(.+)ARRONDISSEMENTS|(.+)ARRONDISSEMENT', 'PARIS', y).strip()))
 .query("ville != 'UAO'")
 )
```

```python
# Correct 75005, 75006
correct_0506 = (pd.concat([df3.loc[[4], :],df3.loc[[4], :]])
 .assign(ids = ['75005EME', '75006EME'])
)
```

```python
criminalite = pd.concat([df3, correct_0506]).query('ids != "75005EME ET 06EME ARRONDISSEMENTS"')
```

```python
criminalite = (criminalite
 .assign(criminalite = lambda x: (x.y16 + x.y17 + x.y18) / 3)
 .loc[:, ['criminalite', 'ville', 'code_postal']]
)
```

```python
population = pd.read_csv('../data/insee-population2015.csv')
```

```python
population = population.iloc[2:, 1:].assign(code_postal=lambda x: x.code_postal.astype('int').astype('str'))
```

```python
crime_normalized = (pd.merge(population, criminalite, left_on='code_postal', right_on='code_postal', how='outer')
 .assign(population = lambda x: x.population / 1000)
 .assign(departement = lambda x: x.code_postal.apply(lambda y: re.search('^\d{2}', y).group(0)))
 .assign(pop_dept = lambda x: x.groupby('departement').population.transform('sum'))
 #.fillna({'criminalite':0})
 .assign(crimin_dept = lambda x: x.groupby('departement').criminalite.transform('sum'))
 .assign(delit_rate = lambda x: np.where(x.criminalite.isna(), x.crimin_dept / x.pop_dept, x.criminalite / x.population))
)
```
